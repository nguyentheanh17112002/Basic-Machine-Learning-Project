### Pyplot Api vs Object-Oriented(OO) API
* Pyplot API $\rightarrow$ Quickly
* OO API $\rightarrow$ Advanced

### Pyplot API

```python
x = np.linspace(0,10,1000)
x
```

```python
plt.plot(x, np.sin(x),color = "red", linestyle ="dashed",label = "sin(x)")
plt.plot(x, np.cos(x),color = "blue",label= "cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

#tile
plt.title("Function of x")

#limit x and y
#plt.xlim([0,8])
#plt.ylim([0,1])

#plt.axis([xmin, xmax, ymin, ymax])
#plt.axis([0,8,-0.5,1])

#plt.axis("tight")


plt.axis("equal")

plt.legend();
```

## OO API

```python
x = [1,2,3,4]
y = [10,20,30,40]

fig , ax = plt.subplots(figsize = (5,5)) # Figure size = Height & Width of the plot

ax.plot(x, y)

ax.set(title = "Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")
```

# Common types of Matplotlib

### Line

```python
x = np.linspace(0,10,100)
x
```

```python
## OO API plot a line chart
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

### Scatter

```python
#Pyplot API
plt.scatter(x , np.exp(x),color = "green");
```

```python
# OO API
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = 100*rng.rand(100)


fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s = sizes , c = color , cmap = "viridis",alpha = 0.3);
fig.colorbar(img1);
```

### Bar
* Vertical
* Horizontal

```python
# python dictionary

soft_drink_prices = {"Coke" : 10,
                    "Pepsi" : 12,
                    "7Up": 17,
                    "Beer":8}
fig , ax = plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values(), color = "green")
ax.set(title = "Price")
```

```python
fig , ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()), color = "green")
ax.set(title = "Price")
```

## Subplots

```python
#option 1:
fig, ((ax1 , ax2) , (ax3 , ax4)) = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,5))
#Plot data to each axis
ax1.plot(x,x/2);
ax1.plot(x,2*x);

ax2.scatter(np.random.random(10),np.random.random(10) );

ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values(), color = "green");

ax4.hist(student_height,color = "grey",bins = 30);
```

```python
#option 2:
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,5))
ax[0,0].plot(x,x/2);
ax[0,0].plot(x,2*x);

ax[0,1].scatter(np.random.random(10),np.random.random(10) );

ax[1,0].bar(soft_drink_prices.keys(),soft_drink_prices.values(), color = "green");

ax[1,1].hist(student_height,color = "grey",bins = 30);
```

# **Project: Data Visualization**

```python
cities = pd.read_csv("./data/california_cities.csv")
cities.head()
```

```python
#Plot using Pyplot API
#Extract latd and longd
lat , lon = cities.latd , cities.longd
population , area = cities.population_total , cities.area_total_km2
plt.style.use("seaborn")
plt.figure(figsize = (8,6))
plt.scatter(lon , lat, c = np.log10(population),cmap = "viridis", s = area , alpha = 0.5);



plt.axis("equal");
plt.xlabel("Longlatitude")
plt.ylabel("Latitude");
plt.colorbar(label = "log$_{10}$(population)")


area_range = [10, 50 , 100 , 300 , 500]

for area in area_range:
    plt.scatter([],[] , s = area , label = str(area) + "km$^2$")
plt.legend(labelspacing = 1 , title = "City area")


plt.title("California cities");
```
